**Data Cleaning And Exploratory Data Analysis**

```python
Titanic.drop(["PassengerId","Name","Ticket","Embarked","Cabin"],axis = 1,inplace = True)
Titanic.head()
```

```python
Titanic.hist(figsize=(13,10),facecolor="orange")
```

```python
sns.barplot(x ='Sex', y ='Survived', data = Titanic,palette ='plasma')
```

```python
sns.countplot(x='Survived',hue='Pclass',data=Titanic)
```

```python
sns.boxplot(x = "Survived",y = "Age", data = Titanic)
```

```python
sns.boxplot(x = "Survived",y = "Parch", data = Titanic)
```

```python
sns.boxplot(x = "Survived",y = "Fare", data = Titanic)
```
